```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = pd.DataFrame([attrition_df['Attrition'], attrition_df['Department']])
y_df = y_df.transpose()
y_df
```

```python
# Create a list of at least 10 column names to use as X data
X_names = ['BusinessTravel', 'DistanceFromHome', 'Education', 'EducationField', 'HourlyRate', 'NumCompaniesWorked', 'PerformanceRating', 'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']


# Create X_df using your selected columns
X_df = attrition_df[X_names]

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_to_encode = ['BusinessTravel', 'DistanceFromHome', 'Education', 'EducationField']


# encoder = OneHotEncoder(drop='first')
# column_transformer = ColumnTransformer(
#     transformers=[
#         ('cat', encoder, X_to_encode)
#     ],
#     remainder='passthrough'
# )
# train_encoded = column_transformer.fit_transform(X_train)
# test_encoded = column_transformer.transform(X_test)

# train_encoded_df = pd.DataFrame(train_encoded, columns=column_transformer.get_feature_names_out())
# test_encoded_df = pd.DataFrame(test_encoded, columns=column_transformer.get_feature_names_out())


df_train_one_hot = pd.get_dummies(X_train, columns=X_to_encode, drop_first=True)
df_test_one_hot = pd.get_dummies(X_test, columns=X_to_encode, drop_first=True)
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(df_train_one_hot)

# Scale the training and testing data
X_train_scaled = scaler.transform(df_train_one_hot)
X_test_scaled = scaler.transform(df_test_one_hot)

X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=df_train_one_hot.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=df_test_one_hot.columns)
```

```python
# Create a OneHotEncoder for the Attrition column


# Fit the encoder to the training data


# Create two new variables by applying the encoder
# to the training and testing data
y_train_one_hot_2 = pd.get_dummies(y_train_one_hot, columns=['Attrition'], dtype='int')
y_test_one_hot_2 = pd.get_dummies(y_test_one_hot, columns=['Attrition'], dtype='int')
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
print(len(X_train_scaled_df.columns))

# # Create the input layer
# model = Sequential()
# model.add(Dense(units=64, activation='relu', input_shape=(45,))
# # Create at least two shared layers
# model.add(Dense(units=48, activation='relu', input_shape=(45,))
# model.add(Dense(units=32, activation='relu', input_shape=(45,))

# Create the input layer
input_layer = Input(shape=(45,))
# Create at least two shared layers
shared_layer_1 = Dense(64, activation='relu')
shared_layer_2 = Dense(32, activation='relu')
shared_output_1 = shared_layer_1(input_layer)
shared_output_2 = shared_layer_2(shared_output_1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer
department_layer = Dense(32, activation='relu')  #name='Department')
# Create the hidden layer
department_output_1 = department_layer(shared_output_2)

# Create the output layer
department_output_layer = Dense(3, activation='softmax', name='Department')
department_output_2 = department_output_layer(department_output_1)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer
attrition_layer = Dense(32, activation='relu')  #name='Attrition')
# Create the hidden layer
attrition_output_1 = attrition_layer(shared_output_2)

# Create the output layer
attrition_output_layer = Dense(1, activation='sigmoid', name='Attrition')
attrition_output_2 = attrition_output_layer(attrition_output_1)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[department_output_2, attrition_output_2])

# Compile the model
model.compile(optimizer='adam', loss={'Department': 'mse', 'Attrition': 'mse'})

# Summarize the model
model.summary()
```

```python
attrition_targets = y_train_one_hot_2[['Attrition_No', 'Attrition_Yes']]
department_targets = y_train_one_hot_2[['Department_Human Resources', 'Department_Research & Development', 'Department_Sales']]

# Train the model
model.fit(
    X_train_scaled_df,
    {'Attrition': attrition_targets, 'Department': department_targets},
    epochs=10,
    batch_size=32
)
```
